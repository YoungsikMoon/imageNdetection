```python
# 폴더 생성 함수 정의
import shutil
import os

# parent_folder에 folder_name이 기존에 없을 경우 새로 생성하는 함수 생성
def make_new_folder(parent_forlder, forder_name):
    path = os.path.join(parent_forlder, forder_name)
    try :
        mode = 0o777
        os.mkdir(path, mode)
    except OSError as error :
        print(error)
```

```python
# 작업 디렉터리 설정 '데이터셋 다운로드 부모 경로에서 작업'
%cd C:\project\alphaco\object\roboflow
current_folder = os.getcwd()
```

```python
# 폴더 체크
path = os.path.join(current_folder, 'train')
os.listdir(path)
```

```python
# 데이터 병합

# 합칠 폴더 리스트
folders = ['test', 'train', 'valid']

# folders에 포함된 폴더 내부 콘텐츠 리스트를 딕셔너리로 저장
content_list = {}
for index, folder in enumerate(folders):
    path = os.path.join(current_folder, folder)
    filenames = os.listdir(path)
    filenames.remove('_darknet.labels')
    content_list[folders[index]] = filenames


content_list
```

```python
# 새롭게 생성할 폴더명 지정
merge_folder = "images"

# 새롭게 생성할 폴더 디렉토리 지정
merge_folder_path = os.path.join(current_folder, merge_folder)

# 새로운 폴더 생성
make_new_folder(current_folder, merge_folder)
```

```python
merge_folder_path
```

```python
# 모든 파일을 수회하며
to_path = merge_folder_path

for sub_dir in content_list: 
    for contents in content_list[sub_dir]: 
        from_path = os.path.join(current_folder, sub_dir, contents)
        # 파일 이동
        shutil.move(from_path, to_path)
```

```python
# 폴더 컨텐츠 개수 출력
len(os.listdir('images'))
```

```python
# mask_data.data 파일 생성
t = open(current_folder+'/mask_data.data', 'w')
```

```python
# 이후 이미지 데이터셋을 저장할 경로 지정
path = current_folder+"/images/"

# test.txt 생성
# w 모드로 text 파일을 생성
f = open(current_folder+'/test.txt', 'w')
for img_fname in content_list['test']:
    if img_fname[-3:]=='jpg':
        # 문자 입력
        f.write(path+img_fname+"\n")
f.close()

# train.txt 생성
# w 모드로 text 파일을 생성
f = open(current_folder+'/train.txt', 'w')
for img_fname in content_list['train']:
    if img_fname[-3:]=='jpg':
        # 문자 입력
        f.write(path+img_fname+"\n")
f.close()

# valid.txt 생성
# w 모드로 text 파일을 생성
f = open(current_folder+'/valid.txt', 'w')
for img_fname in content_list['valid']:
    if img_fname[-3:]=='jpg':
        # 문자 입력
        f.write(path+img_fname+"\n")
f.close()
```
